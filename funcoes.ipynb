```python
def calculadora(a, b, operacao):
    if operacao == 'soma':
        return a + b
    elif operacao == 'subtracao':
        return a - b
    elif operacao == 'multiplicacao':
        return a * b
    elif operacao == 'divisao':
        if b != 0:
            return a / b
        else:
            return "Erro: divisão por zero."
    else:
        return "Operação inválida."
    print(calculadora(a, b, 'soma'))
    print(calculadora(a, b, 'subtracao'))
    print(calculadora(a, b, 'multiplicacao'))
    print(calculadora(a, b, 'divisao'))
    print(calculadora(a, b, 'divisao'))
```

```python
calculadora(15, 10, 'subtracao')
```

```python
def avaliar_aluno(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

    print(avaliar_aluno(nota))
```
